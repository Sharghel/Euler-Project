```python
# Problem 1 :
BORNE = 1000
summ = 0
for i in range(BORNE):
    if i%3 == 0 or i%5 == 0:
        summ += i
print(summ)
```

```python
# Problem 2 :
FibTab = [0, 1]
sum = 0

while FibTab[-1] < 4000000:
    FibTab.append(FibTab[-1]+ FibTab[-2])
for val in FibTab:
    if val%2 == 0:
        sum += val
print(sum)
```

```python
# Problem 3 :
from math import sqrt
import sympy as sp

n = 600851475143
for i in range(1, int(sqrt(n))):
    if n%i == 0 and sp.isprime(i):
        print(i)
```

```python
# Problem 4 :
a = 100
b = 100
s = "0"
tab = []

while (a != 1000) or (b != 999):
    s = str(a*b)
    if s == s[::-1]:
        tab.append(int(s))
    if b == 999:
        a += 1
        b = 100
    else:
        b += 1
print(max(tab))

```

```python
# Problem 6 :
def SumSquare(n):
    summ = 0
    for i in range(1, n+1):
        summ += i**2
    return summ

def SquareSum(n):
    summ = 0
    for i in range(1, n+1):
        summ += i
    return summ**2

def delta(n):
    summ = SumSquare(n)
    squareSumm = SquareSum(n)
    print(squareSumm-summ)

delta(100)
```

```python
# Problem 7 :
import sympy as sp

tab = []
ctp = 1
while len(tab) <= 10002:
    if sp.isprime(ctp):
        tab.append(ctp)
    ctp += 1
print(tab[10000])
```

```python
# Problem 9 :
for a in range(1, 1000):
    for b in range(a + 1, 1000 - a):
        c = 1000 - a - b
        if a**2 + b**2 == c**2:
            print(a*b*c)
```

```python
# Problem 12 :
# 500 diviseur de n
from math import sqrt

def count_divisors(n):
    divisors = 0
    sqrt_n = int(sqrt(n))
    for i in range(1, sqrt_n + 1):
        if n % i == 0:
            divisors += 2  # i and n/i are both divisors
    if sqrt_n * sqrt_n == n:  # If n is a perfect square
        divisors -= 1  # Adjust for the double-counted divisor
    return divisors

# Function to find the first triangular number with over 'limit' divisors
def find_triangular_number_with_divisors(limit):
    n = 1
    triangular = 1
    
    while True:
        # Calculate the next triangular number
        if count_divisors(triangular) > limit:
            return triangular
        n += 1
        triangular += n  # Generate the next triangular number

# Find the first triangular number with more than 500 divisors
result = find_triangular_number_with_divisors(500)
print(f"The first triangular number with over 500 divisors is: {result}")
```

```python
# Problem 18 :
def max_path_minmax(triangle):
    # Starting from the second last row of the triangle, we go upwards.
    for i in range(len(triangle) - 2, -1, -1):
        for j in range(len(triangle[i])):
            # In each move, consider the minimum of the two possible choices
            # to mimic a Minimax step, but we still aim to maximize our result.
            # Normally you'd take the max, but to simulate a minimizing opponent, we use min.
            triangle[i][j] += max(min(triangle[i+1][j], triangle[i+1][j+1]), triangle[i+1][j], triangle[i+1][j+1])
    
    # The top element now contains the maximum total
    return triangle[0][0]

# Triangle input
triangle = [
    [75],
    [95, 64],
    [17, 47, 82],
    [18, 35, 87, 10],
    [20, 4, 82, 47, 65],
    [19, 1, 23, 75, 3, 34],
    [88, 2, 77, 73, 7, 63, 67],
    [99, 65, 4, 28, 6, 16, 70, 92],
    [41, 41, 26, 56, 83, 40, 80, 70, 33],
    [41, 48, 72, 33, 47, 32, 37, 16, 94, 29],
    [53, 71, 44, 65, 25, 43, 91, 52, 97, 51, 14],
    [70, 11, 33, 28, 77, 73, 17, 78, 39, 68, 17, 57],
    [91, 71, 52, 38, 17, 14, 91, 43, 58, 50, 27, 29, 48],
    [63, 66, 4, 68, 89, 53, 67, 30, 73, 16, 69, 87, 40, 31],
    [4, 62, 98, 27, 23, 9, 70, 98, 73, 93, 38, 53, 60, 4, 23]
]

# Call the function and print the result
max_total = max_path_minmax(triangle)
max_total
```

```python
# Problem 19 :
def leapyr(y):
    if y % 400 == 0:
        return True
    elif y % 100 == 0:
        return False
    elif y % 4 == 0:
        return True
    return False

# Initialization
y = 1901
m = 1
day_count = 2  # January 1, 1901 was a Tuesday (day 2 of the week)
sund_1st = 0
m_31 = [1, 3, 5, 7, 8, 10, 12]
m_30 = [4, 6, 9, 11]

# Loop from January 1901 to December 2000
while not (y == 2000 and m == 12):
    
    # Check if the 1st of the current month is a Sunday
    if day_count % 7 == 0:
        sund_1st += 1

    # Move to the next month and adjust day_count by the number of days in the month
    if m == 2:  # February
        if leapyr(y):
            day_count += 29
        else:
            day_count += 28
    elif m in m_31:
        day_count += 31
    elif m in m_30:
        day_count += 30
    
    # Move to the next month
    m += 1
    if m > 12:  # If month exceeds December, move to the next year
        m = 1
        y += 1

# Check December 2000's 1st day after the loop
if day_count % 7 == 0:
    sund_1st += 1

# Output the result
print(sund_1st)
```
